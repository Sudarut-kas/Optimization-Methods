```python
import numpy as np
```

```python
def secant(df, x0, x1, max_iter=100, tol=1e-3):
  for i in range(max_iter):
    y0 = df(x0)
    y1 = df(x1)

    x = (x0*y1 - x1*y0) / (y1 - y0) 
    y = df(x)

    if np.abs(x-x1) / np.abs(x1) < tol:
      break
    
    x0 = x1
    x1 = x
    y0 = y1
    y1 = y

  return x
```
