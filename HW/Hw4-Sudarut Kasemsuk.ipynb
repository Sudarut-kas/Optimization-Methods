#Question 1
(4 pts)
Slove the following "projection on a Euclidean ball" problem via the dual function. The $c, z \in \mathbb{R}^n$ and $r > 0$ are parameters.
$$
\min_{x \in \mathbb{R}^n : \| x-c \| \leq r} \| z-x \|_{2}
$$
Hint: make the target function differentable and rewrite constraints in a more convenient form.

**Solution:** $\min_{\| x-c \|_{2} \leq r } \| z-x \|_{2} $ is equivalent to $\min_{\| x-c \|^{2}_{2} \leq r^{2}} \| z-x \|^{2}_{2} $ \\

Lagrangian function:
\begin{split}
  L(x, \lambda) &= \| z-x \|^{2}_{2} + \lambda (\| x-c \|^{2}_{2} - r^{2}), \lambda \geq 0 \\
 \nabla_{x}L(x, \lambda)  &= -2(z-x) +  2\lambda(x-c) = 0 \\
  x^* &= \frac{1}{1+\lambda}(z+\lambda c)
\end{split}

Dual function:  \\
\begin{split}
  d(\lambda) &= \inf_{x}L(x, \lambda) \\
  &= \| \frac{z(\lambda + 1)-(z+\lambda c)}{ \lambda + 1} \|^{2}_{2} +
  \lambda \| \frac{(z+\lambda c)-c(\lambda + 1)}{ \lambda + 1} \|^{2}_{2} - \lambda r^{2} \\
  &= \frac{\lambda^2}{(\lambda + 1)^2} \| z-c \|^2_2 + \frac{\lambda}{(\lambda + 1)^2} \| z-c \|^2_2 - \lambda r^{2} \\
  &= \frac{\lambda}{(\lambda + 1)} \| z-c \|^2_2 - \lambda r^2 \\
  \nabla d(\lambda) &= \frac{1}{(1+\lambda)^2} \| z-c \|^2_2 - r^2 \\
  \lambda^* &= \frac{\|z-c\|}{r} - 1
\end{split}
So, that 
\begin{split}
x^* &= \frac{r}{\| z-c \|}(z-c+ \frac{\| z-c \|}{r}c) \\
&= \frac{z-c}{\| z-c \|}r + c
\end{split}

$$

$$

v^* = 2\frac{(z,a)-b}{(a,a)}
$$
So that, 
$$
x^* = z - \frac{(z,a)-b}{(a,a)}a

#Question 4
(4pts) Derive (Lagrange) dual function  $d(\cdot),$ and write down the dual problem for the following prima; problem:
$$
\min_{(Ax, x) \leq 1 } (c, x),
$$
where $x, c \in \mathbb{R}^n, c \neq 0, A \in \mathbb{S}^n, A \succ 0.$

**Solution:** \\
Lagrangian funtion: \\
\begin{split}
L(x, \lambda) &= (c,x) + \lambda((Ax,x)-1), \lambda \geq 0 \\
\nabla_{x} L(x, \lambda) &= c + 2Ax\lambda \\
x^* &= \frac{-1}{2\lambda}A^{-1}c
\end{split}

Dual function: \\
\begin{split}
d(\lambda) &= \inf_{x}L(x, \lambda) \\
&= (c,\frac{-1}{2\lambda}A^{-1}c) + \lambda((A(\frac{-1}{2\lambda}A^{-1}c),\frac{-1}{2\lambda}A^{-1}c)-1) \\
&= \frac{1}{4\lambda}(c, A^{-1}c)-\lambda \\
\nabla d(\lambda) &= \frac{1}{4 \lambda}(c, A^{-1}c) - 1 = 0\\
\lambda &= \frac{\sqrt{(c, A^{-1}c)}}{2} 
\end{split}
So that, 

\begin{split}
x^* &= \frac{-1}{2(\frac{\sqrt{(c, A^{-1}c)}}{2} )}A^{-1}c \\
&= \frac{-A^{-1}c}{\sqrt{(c, A^{-1}c)})}
\end{split}

#Question 6
(4 pts) Solve the problem using the gradient with projection method:
$$
\min (c,x) \\
\text{s.t. }(a_i, x) \leq b_i,
$$
where $A \in \mathbb{R}^{(n+1) \times n}, b \in \mathbb{R}^{n+1} x \in \mathbb{R^n}, $ with $i=1, ..., 101; n=100.$

\begin{equation*}
A = 
\begin{bmatrix}
-1 & 0 & \cdots & 0 \\
0 & -1 & 0\cdots & 0 \\
\hdashline
0 & 0  & 0\cdots & -1  \\
1 & 1 & 1\cdots & 1 
\end{bmatrix}
, b = [0, 0, 0, ..., 1]^{T}, c = [1,1,1,...,1]^{T}. 
\end{equation*}

```python
import numpy as np
```

```python
#define A, b, c
def define_A_b_c(n):
  A = np.zeros((n+1,n))
  for i in range(0,n):
    A[i][i] = -1
    A[n][i] = 1

  b = np.zeros((n+1,1))
  b[n] = 1

  c = np.ones((n,1))

  return A, b, c
```

```python
def gradient_with_projection(c, A, b, x0, tol=1e-6, max_iter=1000):
  x = np.array(x0)
  for k in range(max_iter):
    # Compute the gradient of f(x)
    grad_f = c

    # Compute the projected gradient of f(x)
    grad_proj = np.maximum(0, grad_f  + A.T @ np.maximum(0, A @ x - b))
    t = 1
    x_new = x - t * grad_proj

    #Stopping criteria
    if np.linalg.norm(x_new - x) < tol:
      break

    x = x_new
  print(f'Iter: {k+1}')

  return x
```

```python
A, b, c = define_A_b_c(n=4)
x = np.array([0,0,0,1])
x = x.reshape((x.shape[0], 1))
```

```python
gradient_with_projection(c, A, b, x0=x, tol=1e-6, max_iter=1000)
```
