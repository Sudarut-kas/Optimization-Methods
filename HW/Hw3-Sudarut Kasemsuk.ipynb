#Question 1
(1 pt)
Prove that if functions $f(.), g(.)$ are convex, then $h(x) = max{f(x), g(x)}$ is convex as well.


Definition:
A function $f : \mathbb{R}^n \rightarrow \mathbb{R} $ is convex if its domain is a convex set and for all $x, y$ in its domain, and all $\lambda \in [0, 1]$, we have
$$
f(\lambda x + (1-\lambda)y) \leq \lambda f(x) + (1-\lambda)f(y)
$$
Proof:
\begin{split}
  h(\lambda x + (1-\lambda)y) &= \max \{ f(\lambda x + (1-\lambda)y), g(\lambda x + (1-\lambda)y) \} \\
  & \leq \max \{\lambda f(x) + (1-\lambda)f(y), \lambda g(x) + (1-\lambda)f(y) \} \\
  &\leq \lambda \max \{f(x), g(x)\} + (1-\lambda) \max \{f(y), g(y) \} \quad (*)\\
  &= \lambda h(x) + (1-\lambda)h(y)
\end{split}
(*) $\|a+b, c+d\|_{\infty} = \| [a, c] + [b, d] \|_{\infty} \leq \| [a,c] \|_{\infty} + \| [b, d] \|_{\infty} $ \\
From the result, $h(x)$ is convex

Find conjugate function $f^*(y)$ for $f(x) = (Qx, x)$ and $ Q \succ 0$.

The conjugate function $f$ is 
$$
f^*(y) = \sup_{x \in domf} (y^Tx-f(x))
$$

Since $f(x) = (Qx, x) = x^TQ^Tx $ we have,
\begin{split}
f^*(y) &= \sup_{x \in domf} (y^Tx - x^TQ^Tx) \\
&= \sup_{x \in domf} (y^Tx - x^TQx) \quad  (Q \succ 0)
\end{split}

Let $g(x) = y^Tx - x^TQx$
\begin{split}
  \nabla_xg(x) &= \frac{d}{dx}(gx) \\
  &= y - 2Qx \\
\end{split}

Since $y - 2Qx = 0$ so, $x = \frac{y}{2Q}$
\begin{split}
  f^*(y) &= y^T\frac{y}{2Q} - (\frac{y}{2Q})^TQ\frac{y}{2Q} \\

  &= \frac{y^Ty}{2Q} - \frac{y^Ty}{4Q} \quad (Q \succ 0) \\
  &= \frac{y^Ty}{4Q}
\end{split}

Solve one-dimensional problem (approximation by constant)
$$
\min _{z \in \mathbb{R}} \frac{1}{m} \sum^{m}_{i=1}(z - x_{i})^2.
$$

Rewrite function: \\
\begin{split}
  f(z) &= \frac{1}{m} \sum_{i=1}^{m} (z-x_{i})^2 \\
  &= \frac{1}{m}[(z^2-2zx_1+x_1^2)+...+(z^2-2zx_m+x_m^2)] \\
  &= \frac{1}{m}[mz^2 - 2z(x_1+...+x_m)+(x_1^2+...+x_m^2)] \\

\begin{split}

\end{split}
So that, $z = \frac{1}{m}\sum_{i=1}^{m}x_{i}$ \\
Second derivative:
$$
\frac{d(2z - \frac{2}{m} \sum_{i=1}^{m}x_{i})}{dz} = 2 > 0

Let us consider a convex quadratic function:
$$
f(x) = x^TAx,
$$
where $A$ is a symmetric positive definite matrix with controllable $ k = \frac{\lambda_{max}}{\lambda_{min}}$. 
You may consider that $A$ is an $n \times n$ Hilbert matrix (assume $n=5$) given by,
$$
A_{ij} = \frac{1}{i+j-1}, \quad i = 1,2,...,5 \quad j = 1,2,...,5.
$$
1. Analyze the spectrum of the Hessian of the function, find maximum and minimum eigenvalues $L$ and $m$.

```python
from scipy.linalg import hilbert
from numpy import linalg as LA
from sympy import symbols, Eq, solve, log
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

```python
def hessian_matrix(dimension=5):
  #define variables
  x1, x2, x3, x4, x5 = symbols('x1 x2 x3 x4 x5')
  x_list = [x1, x2, x3, x4, x5]
  x = np.matrix([[x1, x2, x3, x4, x5]])

  #calulate function from x.TAx
  A = hilbert(dimension)
  f = np.dot(np.dot(x, A), x.T)
  function = f.item(0)

  #first derivative of function by df/dx1,...,df/dx5
  first_diff = []

  #second derivative of function by df/dx1x1,...,df/dx5x5
  second_diff = []
  for i in range(dimension):
    first_der = function.diff(x_list[i])
    first_diff.append(first_der)
    for j in range(len(x_list)):
      second_der = first_der.diff(x_list[j])
      second_diff.append(second_der)
  
  #construct hessian matrix
  hessian_mat = np.array(second_diff, dtype=np.float64).reshape(5, 5)

  return hessian_mat
```

```python
#calculate hessian matrix from function
hessian_mat = hessian_matrix()

#find eigen values
eigen_value, eigen_vector = LA.eig(hessian_mat)

#min, max of eigen values
print(f' Max eigenvalue: {max(eigen_value)}')
print(f' Min eigenvalue: {min(eigen_value)}')
```

2. Implement the following methods: \\

* Gradient descent with a constant step size, e.g., $\gamma = \frac{1}{L}$ \\

```python
def quadratic_func (A, x):
  #2Ax
  dx = 2* np.dot(A,x)   
  return dx 

def find_gamma():
  hessian_mat = hessian_matrix()
  eigen_value, eigen_vector = LA.eig(hessian_mat)
  L = max(eigen_value)
  gamma = 1 / L 
  return gamma

def func(x, A):
  return x.T @ A @ x 
```

```python
def gradient_descent(x, n, epoch):
  
  A = hilbert(n)
  gamma = find_gamma()
  a = np.array(x)
  ai = []
  ff = []
  dots = []
  k = 1
  calls = 0
  for i in range(epoch):
    a_prev = a
    f = func(a, A)
    ai.append([a, f])
    dots.append(a)
    ff.append(f)
    sk = np.array(quadratic_func(A, a))
    a = a - np.dot(gamma, sk)
    k+=1
    calls += 2

    if np.abs(func(a, A) - func(a_prev, A)) < 1e-6:
      calls += 2
      break

  ai = np.array(ai)
  print(f'iterations : {i+1}')
  print(f'calls : {calls}')
  print(f'at point: {ai[-1,0]} the minimum is: {ai[-1, 1]} ')
  plt.plot(range(i+1), ff)
  plt.ylabel('f(x)')
  plt.xlabel('iteration')
  return dots, ff 
```

```python
x = [0.2, 0.2, 0.4, 0.5, 0.6]
x_grad_ds, f_grad_ds = gradient_descent(x, 5, 1000)
```

  * Gradient descent with a smaller step size (Armijo/Wolfe/Goldstein, etc.)

```python
def gradient_descent_armijo(a, n , alpha=1, beta=0.5, c=1e-4, epoch=2000, tol=1e-5):
  #find gradient descent with armijo condition 
  A = hilbert(n)
  x = np.array(a)
  f_val = func(x, A)
  grad_f_val = quadratic_func(A, x)
  iter = 0
  ai = []
  ff = []
  calls = 2  
  while np.linalg.norm(grad_f_val) > tol and iter < epoch:
    ai.append([x])
    ff.append(f_val)
    iter += 1
    alpha = alpha
    while func(x - alpha * grad_f_val, A) > f_val - c * alpha * np.linalg.norm(grad_f_val)**2:
      alpha *= beta
    x = x - alpha * grad_f_val
    f_val = func(x, A)
    grad_f_val = quadratic_func(A, x)
    calls += 3

  ai = np.array(ai)
  print(f'iterations : {iter}')
  print(f'calls : {calls}')
  print(f'at point: {ai[-1]} the minimum is: {ff[-1]} ')
  #print(f'the minimum is: {ff[-10:]} ')
  plt.plot(range(iter), ff)
  plt.ylabel('f(x)')
  plt.xlabel('iteration')
  return ai, ff
```

```python
x = np.array([0.2, 0.2, 0.4, 0.5, 0.6])
x_grad_ds_ami, f_grad_ds_ami = gradient_descent_armijo(a=x, n=5)
```

```python
def alpha_ls(A, x, d): 

    numerator = np.dot(x, np.dot(A, d))
    denominator = np.dot(d, np.dot(A, d))
    alpha = numerator / denominator

    return alpha
```

```python
def gradient_descent_excat_ls(a, n , epoch=2000, tol=1e-5):
  #find gradient descent with armijo condition 
  A = hilbert(n)
  x = np.array(a)
  f_val = func(x, A)
  grad_f_val = quadratic_func(A, x)
  iter = 0
  ai = []
  ff = []
  calls = 2
  while np.linalg.norm(grad_f_val) > tol and iter < epoch:
    ai.append([x])
    ff.append(f_val)
    iter += 1
    #alpha = find_alpha(x, quadratic_func(A, x))
    alpha = alpha_ls(A, x, grad_f_val)
    x = x - alpha * grad_f_val
    f_val = func(x, A)
    grad_f_val = quadratic_func(A, x)
    calls += 2

  ai = np.array(ai)
  #print(ai[-1])
  print(f'iterations : {iter}')
  print(f'calls : {calls}')
  print(f'at point: {ai[-1]} the minimum is: {ff[-1]} ')
  plt.plot(range(iter), ff)
  plt.ylabel('f(x)')
  plt.xlabel('iteration')
  return ai, ff
```

* Heavy ball method

```python
def grad(A, x):
  #2Ax
  dx = 2* np.dot(A,x)   
  return dx 

def find_alpha_beta():
  hessian_mat = hessian_matrix()
  eigen_value, eigen_vector = LA.eig(hessian_mat)
  L = max(eigen_value)
  M = min(eigen_value)

  # alpha = 4/(np.sqrt(L)+np.sqrt(M)**2)
  # beta = ( (np.sqrt(L) - np.sqrt(M))/(np.sqrt(L)+np.sqrt(M)) )**2

  alpha = 1/L
  beta = 0.9

  return alpha, beta

def func(A, x):
  return x.T @ A @ x 
```

```python
def heavy_ball(x, n=5, epoch=2000):
  
  A = hilbert(n)
  x = np.array(x)

  ls_x = []
  ls_f = []

  alpha, beta = find_alpha_beta()

  x0 = x
  ls_x.append([x0])
  ls_f.append(func(A, x0))

  x = x0 - alpha * grad(A, x0)
  ls_x.append([x])
  ls_f.append(func(A, x))

  x_0 = x0
  x_1 = x
  calls = 3
  for i in range(epoch):
    x = x_1 - alpha * grad(A, x_1) + beta * (x_1 - x_0)
    
    x_0 = x_1
    x_1 = x

    f = func(A, x)
    ls_x.append([x])
    ls_f.append(f)
    calls += 2

    if np.linalg.norm(grad(A, x)) < 1e-5:
      calls += 1
      break

  print(f'iterations : {i}')
  print(f'calls : {calls}')
  print(f'at point: {ls_x[-1]} the minimum is: {ls_f[-1]} ')
  # plt.plot(range(i+2), ls_f)
  # plt.ylabel('f(x)')
  # plt.xlabel('iteration')
  return ls_x, ls_f
```

* Conjugate gradient method

* $\alpha = - \frac{(g^k, p^k)}{(Ap^k, p^k)}$
* $x^{k+1} = x^k + \alpha_k p^k$
* $g^{k+1} = g^k + \alpha_{k}Ap^k $
* $\beta = \frac{(g^{k+1}, Ap^k)}{(Ap^k, p^k)}$
* $p^{k+1} = -g^{k+1} + \beta_{k}p^k$

```python
def conjugate_gradient(x0,tolerance): 
    xk = x0
    A = hilbert(5)
    gfk = np.dot(A, xk)  
    pk = -gfk 
    err = np.linalg.norm(gfk, ord=2) 
    x_l = []
    f_l = []
    num_steps = 0
    calls = 0
    while err > tolerance:
        apk = np.dot(A, pk) 
        gfk2 = np.dot(gfk, gfk)
        
        alpha = gfk2 / np.dot(apk, pk) 
        xk = xk + alpha * pk 
        gfk = gfk + alpha * apk 
        beta = np.dot(gfk, apk) / np.dot(apk, pk) 
        pk = -gfk + beta * pk
        
        num_steps += 1
        err = np.linalg.norm(gfk) 
        x_l.append(xk)
        f = func(A, xk)
        f_l.append(f)
        calls += 1
    return num_steps, func(x_l[-1], A), x_l[-1], x_l ,x_l, f_l , calls

x0 = x = np.array([0.2, 0.2, 0.4, 0.5, 0.6])
num_steps_res,f_res, x_k_res, x_l_res, x_cjg, f_cjd, calls = conjugate_gradient(x0,tolerance=1e-6)
print(f"x: {x_k_res}\niterations: {num_steps_res}")
print(f'calls : {calls}')

  
```

* Fast gradient method

\begin{split}
  y^0 &= x^0 \\
  x^{k+1} &= y^k - \frac{1}{L} \nabla f(y^k) \\
  y^{k+1} &= x^{k+1} + \frac{\sqrt{L}-\sqrt{M}}{\sqrt{L}+\sqrt{M}}(x^{k+1} - x^k)
\end{split}

```python
def grad(A, x):
  #2Ax
  dx = 2* np.dot(A,x)   
  return dx 

def find_LM():
  hessian_mat = hessian_matrix()
  eigen_value, eigen_vector = LA.eig(hessian_mat)
  L = max(eigen_value)
  M = min(eigen_value)
  return L, M

def func(A, x):
  return x.T @ A @ x 
```

```python
def fast_gradient(x, n=5, epoch=2000):
  
  A = hilbert(n)
  x = np.array(x)
  y = x

  L, M = find_LM()

  ls_x = []
  ls_f = []

  ls_x.append([x])
  ls_f.append(func(A, x))
  calls = 1
  for i in range(epoch):
    x_prev = x
    x = y - (1/L)* grad(A, y)
    y = x + (np.sqrt(L)-np.sqrt(M))/(np.sqrt(L)+np.sqrt(M)) * (x - x_prev)

    f = func(A, x)

    ls_x.append([x])
    ls_f.append(f)
    calls += 2

    if np.linalg.norm(grad(A, x)) < 1e-6:
      calls += 1
      break

  # print(f'iterations : {i+2}')
  print(f'at point: {ls_x[-1]} the minimum is: {ls_f[-1]} ')
  print(f'iterations : {i+1}')
  print(f'calls : {calls}')
  # plt.plot(range(i+2), ls_f)
  # plt.ylabel('f(x)')
  # plt.xlabel('iteration')
  return ls_x, ls_f
```

```python
x = [0.2, 0.2, 0.4, 0.5, 0.6]
x_fast_gd, f_fast_gd = fast_gradient(x, n=5, epoch=2000)
```
